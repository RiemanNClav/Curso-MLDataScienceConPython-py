```python
#Estas columnas no sirven de nada, si no las adjunto al dataset original, eliminamos las variables que originaban la variable sex 
#y se añaden nuevo las variables dummy originadas por esto. 
column_name = datos.columns.values.tolist()                #Estos son los nombres de las columnas del dataframe. 
column_name
```

```python
#haremos: 
datos.drop(["sex"], axis = 1) #para borrar una columna del dataframe, axis=1 carga la columna llamada "sex".
```

```python
#si queremos añadir las variables dummy, lo que tenemos que hacer 
pd.concat([datos,dummy_sex], axis = 1) #con esto combinamos, quitamos la columna "sex" y hemos añadido las variables dummy. 
```

```python
#para no hacerlo todo de golpe, solemos hacerlo todo a traves de una funcion. 
def createDummies(df, var_name):
    dummy = pd.get_dummies(df[var_name], prefix = var_name)
    df = df.drop(var_name, axis=1)
    df = pd.concat([df, dummy], axis=1)
    return df
```

```python
#con np.ceil() lo que hacemos es aproximar al siguiente numero entero, el resultado principal. 
#con int() convertimos a numero entero, un resultado decimal. 
k = int(np.ceil(1+np.log2(3333)))
plt.hist(data["Day Calls" ], bins =k ) #El parametro bins son las divisiones, las reparticiones. 
plt.xlabel("Numero de llamas al dia")
plt.ylabel("Frecuencia")
```
